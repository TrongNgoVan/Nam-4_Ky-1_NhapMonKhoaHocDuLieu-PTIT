```python
import networkx as nx
import matplotlib.pyplot as plt

# Hàm tạo đồ thị cho một ngày cụ thể
def create_graph_for_day(shops, creators, products, day):
    G = nx.DiGraph()  # Tạo đồ thị có hướng

    # Thêm các nút shop vào đồ thị
    for shop, data in shops.items():
        G.add_node(shop, type="shop", day=day, color='red', shape='o', **data)  # 'o' là hình tròn

    # Thêm các nút creator vào đồ thị
    for creator, data in creators.items():
        G.add_node(creator, type="creator", day=day, color='blue', shape='s', **data)  # 's' là hình vuông

    # Thêm các nút product vào đồ thị
    for product, data in products.items():
        G.add_node(product, type="product", day=day, color='yellow', shape='^', **data)  # '^' là hình tam giác

    # Thêm các cạnh (quan hệ) giữa các thực thể
    for shop in shops:
        for product in products:
            G.add_edge(shop, product, relation="Bán", day=day)
    for creator in creators:
        for product in products:
            G.add_edge(creator, product, relation="Review", day=day)
        for shop in shops:
            G.add_edge(shop, creator, relation="Hợp đồng", day=day)

    return G

# Dữ liệu mẫu
shops_20241020 = {
    "ErHerdVietNam": {"affiliate_revenue": 1000, "store_revenue": 5000, "avg_price": 200, "total_sales": 50, "self_operated_revenue": 3000},
    "ColorByVietNam": {"affiliate_revenue": 1500, "store_revenue": 7000, "avg_price": 250, "total_sales": 40, "self_operated_revenue": 4000},
}

creators_20241020 = {
    "@tunpham97": {"followers": 10000, "revenue": 2000, "total_sales": 30, "video_revenue": 1000, "live_revenue": 500},
    "@hangkat6668": {"followers": 15000, "revenue": 3000, "total_sales": 25, "video_revenue": 1200, "live_revenue": 710},
    "@hvan_0524": {"followers": 15000, "revenue": 4000, "total_sales": 40, "video_revenue": 1300, "live_revenue": 730},
    "@topgiahn": {"followers": 12000, "revenue": 1500, "total_sales": 20, "video_revenue": 1700, "live_revenue": 750}
}

products_20241020 = {
    "BodyMist": {"revenue": 3000, "total_sales": 20, "avg_price": 150},
    "BU COOK ": {"revenue": 4500, "total_sales": 30, "avg_price": 180},
    "COLORKEY LUMINOUS ": {"revenue": 2400, "total_sales": 45, "avg_price": 170}
}

# Tạo đồ thị
G_20241020 = create_graph_for_day(shops_20241020, creators_20241020, products_20241020, "2024-10-20")

# Bố cục đồ thị
pos = nx.spring_layout(G_20241020)
plt.figure(figsize=(8, 6))

# Vẽ các nút theo loại (hình tròn, vuông, tam giác)
shop_nodes = [node for node, data in G_20241020.nodes(data=True) if data['type'] == 'shop']
creator_nodes = [node for node, data in G_20241020.nodes(data=True) if data['type'] == 'creator']
product_nodes = [node for node, data in G_20241020.nodes(data=True) if data['type'] == 'product']

# Vẽ từng loại nút với hình dạng và màu sắc khác nhau
nx.draw_networkx_nodes(G_20241020, pos, nodelist=shop_nodes, node_color='red', node_shape='o', node_size=1000, label='Shops')
nx.draw_networkx_nodes(G_20241020, pos, nodelist=creator_nodes, node_color='blue', node_shape='s', node_size=1000, label='Creators')
nx.draw_networkx_nodes(G_20241020, pos, nodelist=product_nodes, node_color='yellow', node_shape='^', node_size=1000, label='Products')

# Vẽ các cạnh và nhãn cho đồ thị
nx.draw_networkx_edges(G_20241020, pos, arrowstyle='->', arrowsize=10)
nx.draw_networkx_labels(G_20241020, pos, font_size=10, font_weight='bold')

# Vẽ nhãn cho các cạnh (quan hệ)
edge_labels = nx.get_edge_attributes(G_20241020, 'relation')
nx.draw_networkx_edge_labels(G_20241020, pos, edge_labels=edge_labels)

# Hiển thị chú giải
plt.legend(scatterpoints=1)

plt.title("Mạng lưới đồ thị vào ngày 2024-10-20")
plt.show()
```

```python
import torch
import torch.nn.functional as F
from torch_geometric.nn import GCNConv
from torch_geometric.data import Data

# Giả sử bạn đã có đồ thị G (ở dạng networkx)
# Chuyển đổi đồ thị sang định dạng của PyTorch Geometric
edge_index = torch.tensor(list(G.edges()), dtype=torch.long).t().contiguous()

# Khởi tạo dữ liệu cho mô hình
x = torch.eye(G.number_of_nodes(), dtype=torch.float)  # one-hot encoding cho các node
data = Data(x=x, edge_index=edge_index)

# Định nghĩa GCN model
class GCN(torch.nn.Module):
    def __init__(self):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(G.number_of_nodes(), 16)
        self.conv2 = GCNConv(16, 16)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

# Khởi tạo và huấn luyện mô hình GCN
model = GCN()
output = model(data)
print(output)  # Vector nhúng cho các node
```
